```python
from fbprophet import Prophet
import ccxt
import pandas as pd
```

```python
symbol_data_map = {
    'ETH': 'ETH/USDT',
    'BTC': 'BTC/USDT'
}
```

```python
def get_data(symbol):
    binance = ccxt.binance()
    # each ohlcv candle is a list of [ timestamp, open, high, low, close, volume ]
    return binance.fetchOHLCV(symbol_data_map[symbol], timeframe='5m', params={})
```

```python
exchange_id = 'binance'
exchange_class = getattr(ccxt, exchange_id)
exchange = exchange_class({
    'apiKey': '',
    'secret': '',
})

binance_markets = exchange.load_markets()
```

```python
d = get_data('ETH')
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df[6] = pd.to_datetime(df[0],unit='ms')
```

```python
df_filtered = df[[1,6]]
df_filtered.columns = ['y', 'ds']
df_filtered
```

```python
m = Prophet()
m.fit(df_filtered)
```

```python
future = m.make_future_dataframe(periods=10, freq='1min')
future.tail(20)
```

```python
forecast = m.predict(future)
forecast['ds'] = forecast['ds'].dt.tz_localize('utc').dt.tz_convert('US/Pacific')

predictions = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
predictions.head(10)
```

```python
predictions.tail(10)
```

```python
fig1 = m.plot(forecast)
```
