```python
import pandas as pd
import pandas_datareader as web
import datetime as dt
```

```python
COUNTRIES = {
    'USA': {
        'suffix': '',
        'currency': 'USD'
    },
    'IND': {
        'suffix': '.NS',
        'currency': 'INR'
    }
}

def get_stock_data(stocks, start, end, country):
    
    if isinstance(stocks, list):
        pass
    elif isinstance(stocks, str):
        stocks = [stocks]
    else:
        return "Invalid Ticker exception"
    
    suffixed_ticker = [stock + COUNTRIES[country]['suffix'] for stock in stocks]
    print(suffixed_ticker)
    
    d = web.DataReader(suffixed_ticker, 'yahoo', start, end)
    d[len(d)-10:]['Close']
    print(d)
    
    
    
stocks = ['BA', 'AMD']
stocks_ind = ['SBIN', 'RELIANCE']

INDIA_ISO = 'IND'
USA_ISO = 'USA'


start = dt.datetime(2018, 1, 1)
end = dt.datetime.now()


get_stock_data(stocks, start, end, USA_ISO)
    
    
```
